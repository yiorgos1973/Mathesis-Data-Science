### Ε1: Εισόδημα και Αποτελέσματα των Τεστ

Θα διερευνήσετε τη σχέση μεταξύ του εισοδήματος σε κάθε χώρα και των αποτελεσμάτων των τεστ TIMSS και PIRLS, όπως στο 4.1 και την εικόνα 5 της μελέτης. Συγκεκριμένα, θα διερευνήσετε μέσω παλινδρόμησης τη σχέση μεταξύ του λογαρίθμου του εισοδήματος και του σκορ TIMMS, και του λογαρίθμου του εισοδήματος και του σκορ PIRLS. Αφού φτιάξετε  τα σχετικά μοντέλα και σχολιάσετε τα αποτελέσματά σας, θα πρέπει να παράξετε διαγράμματα των δεδομένων αντίστοιχα με τα παρακάτω.

Για την απάντηση στην ερώτηση θα χρειαστείτε και το αρχείο `WDI_data.csv`, το οποίο περιέχει δεδομένα εισοδήματος ανά χώρα από την Παγκόσμια Τράπεζα.

![GDP (PPP) and TIMSS](gdp_timss.svg)

![GDP (PPP) and PIRLS](gdp_pirls.svg)

```python
wdi = pd.read_csv("WDI_data.csv", na_values=['...', ' ', '', '..'])
wdi.head() # load wdi
```

```python
gdp = wdi[wdi['Series Code'] == 'NY.GDP.PCAP.PP.KD'][['Country Code', '2015 [YR2015]']].dropna()
gdp.sample(10) # extract GDP per Country
```

```python
df = country_scores.merge(gdp, on='Country Code')
df['GDP_log'] = np.log(df['2015 [YR2015]'])
df.head() # completing data source for models
```

```python
plt.style.use('bmh')
line_kws = {'lw':2, 'color':'red', 'label':'Regression'}
f, (ax1, ax2) = plt.subplots(1, 2, sharex=True, figsize=(21, 6))
x, y1, y2 = 'GDP_log', 'Median Math Score (TIMSS Scale, 4th Grade)', 'Median Reading Score (PIRLS Scale, 4th Grade)'
sns.regplot(x=x, y=y1, data=df[[x, y1]], ax=ax1, line_kws=line_kws, marker='^', label='Country')
sns.regplot(x=x, y=y2, data=df[[x, y2]], ax=ax2, line_kws=line_kws, marker='*', label='Country')
ax1.set_ylabel('TIMSS')
ax2.set_ylabel('PIRLS')
ax1.set_xlabel('Log(GDP per Capita 2015)')
ax2.set_xlabel('Log(GDP per Capita 2015)')
ax1.legend(fancybox=True, shadow=True)
ax2.legend(fancybox=True, shadow=True);
```

```python
from sklearn.linear_model import LinearRegression
```

<h3>TIMSS</h3>

Για ευκολία ανάλυσης θα γίνει αντικατάσταση των τιμών που λείπουν με τον μεσο όρο. Η αντικατάσταση αυτή θα αφαιρεσει 0.1 περιπου στο R2.

```python
timss_df = df[['GDP_log', 'Median Math Score (TIMSS Scale, 4th Grade)']].fillna(df[['GDP_log', 'Median Math Score (TIMSS Scale, 4th Grade)']].mean())
X_timss = timss_df['GDP_log'].values.reshape(-1, 1)
y_timss = timss_df['Median Math Score (TIMSS Scale, 4th Grade)'].values
lr_timss = LinearRegression().fit(X_timss, y_timss)
score = lr_timss.score(X_timss, y_timss)
coef_timss = lr_timss.coef_[0]
intercept_timss = lr_timss.intercept_
print(f"R2: {score:1.4f}, coef: {coef_timss:1.4f}, intercept:{intercept_timss:1.4f}")
regression_timss = lr_timss.predict(X_timss)
resid_timss = y_timss - regression_timss

plt.figure(figsize=(8,4))
plt.hist(resid_timss, bins=8)
plt.title("Residual Histogramm TIMSS");
```

<h3>PIRLS</h3>
Για ευκολία ανάλυσης θα γίνει αντικατάσταση των τιμών που λείπουν με τον μεσο όρο όπως πιο πάνω στη ΤΙΜSS. Η αντικατάσταση αυτή θα έχει μια επιρροή κι εδώ στο R2 περίπου +/- 0.1

```python
pirls_df = df[['GDP_log', 'Median Reading Score (PIRLS Scale, 4th Grade)']].fillna(df[['GDP_log', 'Median Reading Score (PIRLS Scale, 4th Grade)']].mean())
X_pirls = pirls_df['GDP_log'].values.reshape(-1, 1)
y_pirls = pirls_df['Median Reading Score (PIRLS Scale, 4th Grade)'].values
lr_pirls = LinearRegression().fit(X_pirls, y_pirls)
score = lr_pirls.score(X_pirls, y_pirls)
coef_pirls = lr_pirls.coef_[0]
intercept_pirls = lr_pirls.intercept_
print(f"R2: {score:1.4f}, coef: {coef_pirls:1.4f}, intercept:{intercept_pirls:1.4f}")
regression_pirls = lr_pirls.predict(X_pirls)
resid_pirls = y_pirls - regression_pirls

plt.figure(figsize=(8,4))
plt.hist(resid_pirls, bins=12)
plt.title("Residual Histogramm PIRLS");
```

<h3>Σχόλια</h3>

Και στα δύο μοντέλα σύγκρισης του λογάριθμου του εισόδηματος με τις μαθησιακές ικανότητες φαίνεται να υπάρχει μια στατιστικά σημαντική συσχέτιση ωστόσο υπάρχουν φαίνεται και άλλοι σημαντικοί παράγοντες. Οι συντελεστές R2 είναι κι οι δύο λιγότερο από 0.50, τα νούμερα αυτά όμως από όσο έχει δείξει η εμπειρία  στις ανθρωπιστικές επιστήμες πρέπει να θεωρούνται σημαντικά.

Όσον αφορά την ικανότητα στα μαθηματικά (ΤΙΜSS) φαίνεται το εισόδημα να παίζει <b>λιγότερο</b> σημαντικό ρόλο σε σύγκριση με την ανάγνωση. Ο συντελεστής της παλινδρόμησης για την PIRLS ειναι περίπου 42 ενω για την ΤΙΜΜS λιγότερο από 44. Επιπλέον ο R2 για το μοντέλο της PIRLS  είναι 0.43 δηλαδή ενω για την  TIMMS 0.36 που σημαίνει ότι το εισόδημα εξηγεί μεγαλύτερη ποικιλότητα για την ανάγνωση από ό,τι για τα μαθηματικά ωστόσο η χαμηλότερη τιμή στην κλιση της παλινδρόμησης δείχνει ότι η ικανότητα στην ανάγνωση περιέχει περισσότερα μη γραμμικά στοιχεία  

Το ιστόγραμμα των υπόλοιπων φαινεται να είναι πιο κοντά στην κανονική κατανομή για τα μαθηματικά από ότι για τα ανάγνωση επιβεβαιώνοντας την ορθότητα των κλισεων της παλινδρόμησης. Ειδικά για την PIRLS φαίνεται μια ομάδα αρνητικών υπολοίπων (-50 έως -100) τα οποία θα πρέπει να εξηγηθούν από κάποια άλλη μεταβλητή.

```python
df_school = df.merge(school2010, right_on='WBcode', left_on='Country Code')
df_school.dropna(subset=['yr_sch'], inplace=True) # we keep the data where yr_sc is present since 
```

```python
source = df_school[['GDP_log', 'yr_sch']]
X_school = source.GDP_log.values.reshape(-1, 1)
y_school = source.yr_sch.values
lr_school = LinearRegression().fit(X_school, y_school)
regression_school = lr_school.predict(X_school)
resid_school = y_school - regression_school
reg_str = f'years_school = {lr_school.coef_[0]:1.4f} * gdp_log {lr_school.intercept_:1.4f}'
score_school = lr_school.score(X_school, y_school)
print("Regression:", reg_str,f'\tScore: R2 ={score_school: 1.4f}', end='\n\n')
```

```python
f, (ax1, ax2) = plt.subplots(ncols=2, figsize=(18, 4))
ax1.hist(resid_school, bins=12,)
ax1.set_title('Histogram of Residuals')
ax1.set_xlabel("Residual")
ax1.set_ylabel("Frequency")
sns.regplot(x=X_school, 
            y=y_school, 
            label='Country', 
            line_kws = {'lw':2, 'color':'red', 'label':'Regression'},
            ax=ax2)
#ax2.legend(mode='expand', loc='upper right', fontsize='x-large', bbox_to_anchor=(0.2, 1.15), shadow=False, ncol=2);
ax2.set_title("Regression to Variability")
ax2.set_ylabel("Years of School")
ax2.set_xlabel("Log(GDP per Capital)")
ax2.legend(fancybox=True, shadow=True)
plt.tight_layout;
```

<h4>Note:</h4>κάντε μια παλινδρόμηση μεταξύ των υπολοίπων TIMMS και των υπολοίπων των ετών εκπαίδευσης και μια παλινδρόμηση μεταξύ των υπολοίπων PIRLS και των υπολοίπων των ετών εκπαίδευσης. Δείξτε τα αποτελέσματα και φτιάξτε διαγράμματα

```python
lr_resid_timms = LinearRegression().fit(resid_school.reshape(-1, 1), df_school['resid_timss'].values)
lr_resid_pirls = LinearRegression().fit(resid_school.reshape(-1, 1), df_school['resid_pirls'].values)
print(lr_resid_pirls.coef_, lr_resid_timms.coef_)
```

```python
def results_plot(model, X, y, xlabel='', ylabel='', utitle=''):
    """"""
    regression = model.predict(X)
    residuals = y - regression
    r2 = round(model.score(X, y), 2)

    f, (ax1, ax2) = plt.subplots(ncols=2, figsize=(18, 6))
    ax1.hist(residuals, bins=12, label=r2)
    ax1.set_title('Histogram of Residuals')
    ax1.set_xlabel("Residual")
    ax1.set_ylabel("Frequency")
    ax1.legend([f'R2: {r2}'], fontsize='xx-large')
    sns.regplot(x=X, 
                y=y, 
                label='Country', 
                line_kws = {'lw':2, 'color':'red', 'label':'Regression'},
                ax=ax2)
    ax2.set_title(f"y = {model.coef_[0]:1.3f} * x + {model.intercept_:1.3f}")
    ax2.set_ylabel(ylabel)
    ax2.set_xlabel(xlabel)
    ax2.legend(fancybox=True, shadow=True)
    plt.suptitle(utitle, fontsize='xx-large')
    plt.tight_layout()
    return f;
```

```python
f_timms = results_plot(lr_resid_timms, resid_school.reshape(-1,1), df_school['resid_timss'].values, 
                       utitle="Residual TIMSS VS Residual School Years",
                       xlabel='Residualized Years of Schooling',
                       ylabel='Residualized TIMMS')
```

```python
f_pirls = results_plot(lr_resid_pirls, resid_school.reshape(-1,1), df_school['resid_pirls'].values, 
                       utitle="Residual PIRLS VS Residual School Years",
                       xlabel='Residualized Years of Schooling',
                       ylabel='Residualized PIRLS')
```

## Κώδικας Τιμής

Συνειδητοποιείτε ότι αυτή είναι ατομική εργασία και θα πρέπει να εργαστείτε μόνοι σας. Μπορείτε βεβαίως να αναζητήσετε βοήθεια στο διαδίκτυο, αλλά αυτή θα πρέπει να αφορά τη χρήση των εργαλείων και ιδιωμάτων, δεν θα πρέπει να αναζητείτε την απάντηση στα ερωτήματα της εργασίας. Μπορείτε να συζητήσετε με άλλους που παρακολουθούν το μάθημα προκειμένου να λύσετε μεταξύ σας απορίες κατανόησης, αλλά όχι μεθόδους επίλυσης. Αναλογιστείτε ότι σκοπός του μαθήματος είναι να αποκομίσετε κάποιες γνώσεις, και όχι να ξοδέψετε απλώς τον χρόνο και τα χρήματά σας.
