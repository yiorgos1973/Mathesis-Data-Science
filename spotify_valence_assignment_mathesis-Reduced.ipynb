```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LassoLarsCV, LassoLarsIC, RidgeCV, LinearRegression
from sklearn.ensemble import RandomForestRegressor

from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.preprocessing import StandardScaler 

from sklearn.metrics import mean_absolute_error

from pprint import pprint


def train_model(estimator, X_train, y_train, X_valid, y_valid, estimator_kwargs={}, fit_kwargs={}):
    """Train and evaluate a model."""
    model = estimator(**estimator_kwargs).fit(X_train, y_train, **fit_kwargs)
    y_pred_train = model.predict(X_train)
    y_pred_valid = model.predict(X_valid)
    residuals = y_valid - y_pred_valid
    mae_train = mean_absolute_error(y_train, y_pred_train)
    mae_test = mean_absolute_error(y_valid, y_pred_valid)
    
    plt.figure(figsize=(6,4))
    plt.hist(residuals, bins=12)
    plt.xlabel("Residual")
    plt.ylabel("Frequency")
    plt.title("Residual Histogram")
    return {'model':model, 'MAE_train': mae_train, 'MAE_test': mae_test}
```

```python
# load data without some features playing no role
data = pd.read_csv('data_spotify_valence.csv', index_col=[0, 'song_id']).drop(columns="type	uri	track_href	analysis_url".split())
data.sample(10)
```

<h3> E2. Deep Learning</h3>

In this project, a deep learning recipee will be applied that will also include regularizer and dropout. Also the validation set will be used to implement an early stopping when the model stops being improved. We are going to use a very small batch size since we do not have much data and smaller batch sizes is proven to work better in Deep Learning. 

```python
from tensorflow import keras
from keras import models, layers, regularizers, callbacks 

def model_recipee(train_data, l2=0.001, activation='relu'): 
    model = models.Sequential() 
    model.add(layers.Dense(16, kernel_regularizer=regularizers.l2(l2), activation=activation, 
                           input_shape=(train_data.shape[1],)))
    model.add(layers.Dropout(0.15,))
    model.add(layers.Dense(32, kernel_regularizer=regularizers.l2(l2), activation=activation)) 
    model.add(layers.Dropout(0.15,))
    model.add(layers.Dense(64, kernel_regularizer=regularizers.l2(l2), activation=activation))
    model.add(layers.Dense(1))
    model.compile(optimizer='adam', loss='mae', metrics=['mae']) 
    return model



def plot_results(history):
    #dict_keys(['val_loss', 'val_mean_absolute_error', 'loss', 'mean_absolute_error'])
    
    keys = history.history.keys()
    val_metric_name = [k for k in keys if k.startswith("val") and "loss" not in k][0]
    train_metric_name = val_metric_name.replace("val_", "")
    
    train_metric = history.history[train_metric_name]
    train_loss = history.history['loss']
    val_metric= history.history[val_metric_name]   
    val_loss= history.history['val_loss']
    
    epochs = range(1, len(train_metric)+1)
    
    plt.figure()
    plt.plot(epochs, train_metric, 'r', label='Training Metric')
    plt.plot(epochs, val_metric, 'b', label='Validation Metric')   
    plt.title("Training & Validation {}".format(train_metric_name))
    plt.xlabel("epoch")
    plt.ylabel(train_metric_name)
    plt.legend()
    plt.show()
    
    plt.figure()
    plt.plot(epochs, train_loss, 'r', label='Training Loss')
    plt.plot(epochs, val_loss, 'b', label='Validation Loss')
    plt.title("Training & Validation Loss")
    plt.xlabel("epoch")
    plt.ylabel("model loss")
    plt.legend()
    plt.show()   
    return True
```

```python
callback = callbacks.EarlyStopping(monitor='loss', patience=100, restore_best_weights=True)
```

```python
model = model_recipee(x_train, l2=0.0)
```
